```python
import pandas as pd

# Dataset base
datos = {
    'Nombre':  ['Ana', 'Bob', 'Clara', 'Diego', 'Eva'],
    'Edad':    [25, 30, 22, None, 28],
    'Ciudad':  ['Madrid', 'Lima', 'Bogotá', 'Medellín', None],
    'Ingreso': [3000, 4500, 2800, 5000, None]
}

df = pd.DataFrame(datos)
df.to_csv('actividad_semana5.csv', index=False)

# 1) Series — crear y operar
serie_lista = pd.Series([10, 20, 30, 40])
serie_dicc = pd.Series({'a': 1, 'b': 2, 'c': 3})
print("Acceso por índice:", serie_lista[2])       # 30
print("Acceso por clave:", serie_dicc['b'])       # 2
serie_lista[1] = 99
print("Serie modificada:", serie_lista)
print("Serie multiplicada:", serie_lista * 2)

# 2) DataFrame — crear y explorar
df_peq = pd.DataFrame({
    'Producto': ['A', 'B', 'C'],
    'Precio': [100, 200, 150],
    'Stock': [10, 5, 8]
})
print(df_peq.dtypes)

# 3) Inspeccionar y explorar
print(df.head())
print(df.tail())
print(df.info())
print("Shape:", df.shape)
print(df.describe())

# 4) Acceder a columnas y filas
print(df['Nombre'])          # columna como Series
print(df.loc[1])             # fila con índice 1
print(df.iloc[1])            # fila con posición 1
print(df.loc[1, 'Edad'])     # valor específico

# 5) Operaciones básicas
df['Edad'] = df['Edad'].fillna(0) + 5
df['Ingreso_anual'] = df['Ingreso'].fillna(0) * 12
print(df)

# 6) Filtrar datos
print(df[df['Edad'] > 30])
print(df[df['Ciudad'].isin(['Madrid', 'Lima'])])

# 7) Manejo de valores faltantes
print(df.isna())
print(df.isna().sum())
df['Edad'] = df['Edad'].fillna(0)
df['Ciudad'] = df['Ciudad'].fillna('Desconocido')
df['Ingreso'] = df['Ingreso'].fillna(df['Ingreso'].median())
print(df)

# 8) Leer y guardar datos
df_leido = pd.read_csv('actividad_semana5.csv')
print(df_leido.head())
df[['Nombre', 'Edad', 'Ciudad']].to_csv('actividad_semana5_reducido.csv', index=False)

# 9) Ejercicio integrado
df['Edad'] = df['Edad'] + 5
df_filtrado = df[df['Ingreso_anual'] > 36000].sort_values(by='Ingreso_anual', ascending=False)
df_filtrado.to_csv('personas_filtradas.csv', index=False)
print(df_filtrado)
```
